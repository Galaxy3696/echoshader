```python
import xarray as xr
import matplotlib.pyplot as plt
```

```python
ds = xr.open_mfdataset('sample_Sv/*_Sv.nc')
```

```python
plt.figure(figsize = (20, 5))
ds.Sv.isel(frequency=0).T.plot(yincrease=False)
```
